```python
def maxmins(datain, datafin,micro):
    '''
    Tem que ter a lista de regioes jah criada com o nome regioes
    As datasin e datafin entram como strings e.g.('2002249' e '2003041')
    '''
    lminimo = []
    lmaximo = []
    for regiao in regioes:
        locais_minimos = np.argmin(dados.loc[datain:datafin,micro], axis=0)
        locais_maximos = np.argmax(dados.loc[datain:datafin,micro], axis=0)
        lminimo.append(locais_minimos)
        lmaximo.append(locais_maximos)
        lmaximo.sort()
        lminimo.sort()
    vminimo = np.min(dados.loc[datain:datafin,:])
    vmaximo = np.min(dados.loc[datain:datafin,:])
    return zip(micro,lminimo,vminimo,lmaximo,vmaximo)
```
