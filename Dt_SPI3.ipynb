```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from pandas import Series
import bottleneck as bn
from scipy import stats

%pylab inline
```

```python
spi3 = pd.read_csv('dados/spi3_meso_mean_std.csv', converters={'data':str})
```

```python
spi3.columns
```

```python
spi3['date'] = pd.to_datetime(spi3['data'], format='%Y%m')
```

```python
spi3 = spi3.set_index(spi3.date)
```

```python
x = pd.to_datetime(spi3['date'], format='%Y%j')
```

```python
m2 = spi3['m4102'].tolist()
m3 = spi3['m4103'].tolist()
m4 = spi3['m4104'].tolist()
m6 = spi3['m4106'].tolist()
s2 = spi3['s4102'].tolist()
s3 = spi3['s4103'].tolist()
s4 = spi3['s4104'].tolist()
s6 = spi3['s4106'].tolist()
```

```python
#Media movel
w = 3
m2 = bn.move_mean(m2, window=w)
m3 = bn.move_mean(m3, window=w)
m4 = bn.move_mean(m4, window=w)
m6 = bn.move_mean(m6, window=w)
```

```python
xl = w +1
xu = len(x)-1
loc = 3
lw = 3
```

```python
## Aqui é a figura!!!
#########################
figura = plt.figure(figsize=(15, 5.5))
# regiao 4102
plt.errorbar(x,m2,fmt='-', color='b', linewidth=lw, yerr=s2, errorevery=4, capsize=8, 
             elinewidth=3, alpha=0.6, zorder=3, barsabove=True, label='Centro Ocidental')
#plt.plot(x,mm2, marker='d',color='b', alpha=0.5, markersize=12, zorder=5)

#regiao 4103
plt.errorbar(x,m3,fmt='-', color='g', linewidth=lw, yerr=s3, errorevery=4, capsize=8,
             elinewidth=3, alpha=0.6, zorder=3, barsabove=True, label='Norte Central')
#plt.plot(x,mm3, marker='d',color='g', alpha=0.5, markersize=12, zorder=5)

#regiao 4104
plt.errorbar(x,m4,fmt='-', color='r',linewidth=lw, yerr=s4, errorevery=4, capsize=8,
             elinewidth=3, alpha=0.6, zorder=3, barsabove=True, label='Norte Pioneiro')
#plt.plot(x,mm4, marker='d', color='r', alpha=0.5, markersize=12, zorder=5)

#regiao 4106
plt.errorbar(x,m6,fmt='-', color='c', linewidth=lw, yerr=s6, errorevery=4, capsize=8,
             elinewidth=3, alpha=0.6, zorder=3, barsabove=True, label='Oeste')
#plt.plot(x,mm6, marker='d', color='c', alpha=0.5, markersize=12, zorder=5)


###propriedades gerais do gráfico
#plt.title('name',y=0.89, fontsize=16, color = 'k') #backgroundcolor = '#FFFFFF'
plt.grid(True, color = '0.75', linestyle='--', linewidth=0.5, zorder=1)
#plt.gca().patch.set_facecolor('0.97') #não entendi, mas serviu bem #F0F0F5 #F5F5F5 OOOOOUUUU 0 a 1
plt.ylabel('SPI3', fontsize=18)
#plt.ylim(-1.25,1.25)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14, rotation=30)
plt.xlim(x[xl], x[xu])
plt.axhline(0, linewidth=1, color='k')
plt.axhline(0.5, linewidth=1, alpha=0.5, color='k')
plt.axhline(-0.5, linewidth=1, alpha=0.5, color='k')
plt.legend(loc=loc, fancybox=True, prop={'size':12})
plt.show()
folder = '/home/denis/Copy/DissDOCS/dissertation/Figures/'

#plt.savefig(folder + 'Fspi3_mm' + str(w) + '_' + str(xl) + '-' + str(xu) + '.pdf', bbox_inches='tight', format='pdf')
```
